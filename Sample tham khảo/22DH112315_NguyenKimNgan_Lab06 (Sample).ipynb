```python
# Bài 2
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Bước 1
who_covid_data=pd.read_csv('WHO-COVID-19-global.csv')
headers=list(who_covid_data)
print(headers)
```

```python
# Bước 2
every_nth_days=1
country_df= who_covid_data[who_covid_data[' Country']=='Brazil']
info=['cases']
for info_type in info:
    print(info_type)
    brazil_df=pd.DataFrame(country_df)
    print(brazil_df.head(10))
    brazil_df.plot()
    plt.show()
```

```python
# Bước 3
every_nth_days=1
country_df= who_covid_data[who_covid_data[' Country']=='Brazil']
info=['cases']
for info_type in info:
    fig, ax1=plt.subplots(figsize=(20,10), dpi=50)
    color='tab:blue'
    ax1.set_xlabel('days', fontsize='large',fontweight='bold')
    ax1.bar(country_df['Date_reported'][::every_nth_days],country_df[' New_{}'.format(info_type)][::every_nth_days],color=color,linewidth=20,label='New {}'.format(info_type))
    ax1.tick_params(axis='y', labelcolor=color, labelsize='large')
    #bo sung
    ax2=ax1.twinx()
    
    color='tab:red'
    ax2.plot(country_df['Date_reported'][::every_nth_days],country_df[' Cumulative_{}'.format(info_type)][::every_nth_days], color=color, label='Cumulative {}'.format(info_type), linewidth=3)    
    ax2.tick_params(axis='y', labelcolor=color, labelsize=20)    
    #end bo sung
    for idx, t in enumerate (ax1.get_xticklabels()):
        t.set_fontsize(11)
        t.set_fontweight('bold')
        if (idx % 5)!=0:
            t.set_visible(False)
    plt.ticklabel_format(style='plain',axis='y')
    plt.xticks(rotation=90)
    dates=country_df['Date_reported'].unique()
    start_date=dates[0]
    end_date=dates[len(dates)-1]
    
    plt.title('{} (from {} to {})'.format('Brazil',start_date, end_date),fontsize=30,fontweight='bold')
    fig.tight_layout()
    ax1.legend(loc='upper left',fontsize=20)
    ax2.legend(loc='upper right',fontsize=20)
    plt.setp(ax1.xaxis.get_majorticklabels(),rotation=45)
    
    plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Bài 3
# Chọn cột dữ liệu cần phân tích và sắp xếp dữ liệu theo ngày
country_df['Date'] = pd.to_datetime(country_df['Date_reported'])
country_df.set_index('Date', inplace=True)

# Lọc dữ liệu từ ngày 256 đến ngày 316
start_date = country_df.index[255]  # Ngày 256 (lưu ý chỉ số bắt đầu từ 0)
end_date = country_df.index[315]    # Ngày 316
data = country_df.loc[start_date:end_date][' New_cases']

# Tạo dữ liệu cho mô hình
X = []
y = []

for i in range(len(data) - 7):
    X.append(data.iloc[i:i+7].values.flatten())
    y.append(data.iloc[i+7])

X = pd.DataFrame(X, columns=[f'Day_{i+1}' for i in range(7)])
y = pd.Series(y, name='Day_8')

# Phân chia dữ liệu thành tập huấn luyện và tập kiểm tra
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Xây dựng mô hình hồi quy tuyến tính
model = LinearRegression()
model.fit(X_train, y_train)

# Dự đoán số ca cho tập kiểm tra
y_pred = model.predict(X_test)

# Đánh giá mô hình
print(f"Mean Squared Error: {mean_squared_error(y_test, y_pred)}")

# Dự đoán số ca cho ngày tiếp theo
latest_data = X.iloc[-1, :].values.reshape(1, -1)  # Lấy dữ liệu 7 ngày gần nhất
next_day_prediction = model.predict(latest_data)
print(f"Dự đoán số ca cho ngày thứ 317: {next_day_prediction[0]}")

# Dự đoán số ca trong nhiều ngày tiếp theo
num_days = 10
future_predictions = []

for _ in range(num_days):
    next_day_prediction = model.predict(latest_data)
    future_predictions.append(next_day_prediction[0])
    # Cập nhật dữ liệu để dự đoán cho ngày tiếp theo
    latest_data = np.roll(latest_data, shift=-1, axis=1)
    latest_data[0, -1] = next_day_prediction

# Tạo dữ liệu cho dự đoán
predicted_cases = []

for i in range(len(data) - 7):
    X_test = data.iloc[i:i+7].values.flatten().reshape(1, -1)
    prediction = model.predict(X_test)
    predicted_cases.append(prediction[0])

# Chọn các ngày từ ngày 263 đến ngày 316 (tương ứng với dữ liệu dự đoán)
date_range = pd.date_range(start=start_date + pd.DateOffset(days=7), end=end_date)

# Vẽ đồ thị số ca thực tế và dự đoán
plt.figure(figsize=(14, 7))
plt.plot(data.index, data, 'b-', label='Số ca thực tế')  # Đường liền cho dữ liệu thực tế
plt.plot(date_range, predicted_cases, 'r--', label='Dự đoán số ca')  # Đường đứt nét cho dự đoán

plt.title('Brazil from day 256 to 316')
plt.xlabel('Ngày')
plt.ylabel('Số ca')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Bài 5
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Đọc dữ liệu từ file CSV
df = pd.read_csv('Salary_dataset.csv')

# Hiển thị 5 dòng đầu tiên của dữ liệu để kiểm tra
print(df.head())

# Loại bỏ các giá trị rỗng
df.dropna(inplace=True)

# Loại bỏ các giá trị trùng lặp
df.drop_duplicates(inplace=True)

# Hiển thị số lượng records còn lại
print(f"Số lượng records còn lại: {len(df)}")

# Hiển thị dữ liệu cột “salary” trên biểu đồ
plt.figure(figsize=(10, 5))
plt.hist(df['Salary'], bins=20, edgecolor='black')
plt.title('Phân phối của Salary')
plt.xlabel('Salary')
plt.ylabel('Số lượng')
plt.grid(True)
plt.show()

# Hiển thị dữ liệu cả 2 cột trên cùng 1 biểu đồ
plt.figure(figsize=(10, 5))
plt.scatter(df['YearsExperience'], df['Salary'], color='blue', label='Salary vs YearsExperience')
plt.title('Mối quan hệ giữa YearsExperience và Salary')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.legend()
plt.grid(True)
plt.show()

# Chia dữ liệu thành tập huấn luyện và tập kiểm tra
X = df[['YearsExperience']]
y = df['Salary']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=0)

# Xây dựng mô hình hồi quy tuyến tính
model = LinearRegression()
model.fit(X_train, y_train)

# Dự đoán kết quả cho tập kiểm tra
y_pred = model.predict(X_test)

# Đánh giá mô hình
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")

# Hiển thị các điểm dữ liệu và đường hồi quy
plt.figure(figsize=(10, 5))
plt.scatter(X_test, y_test, color='blue', label='Dữ liệu kiểm tra')
plt.plot(X_test, y_pred, color='red', linewidth=2, label='Đường hồi quy')
plt.title('Dự đoán Salary dựa trên YearsExperience')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.legend()
plt.grid(True)
plt.show()
```
