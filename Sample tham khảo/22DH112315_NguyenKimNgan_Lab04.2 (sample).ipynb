```python
#Bản _Real
from collections import deque, defaultdict

def read_graph_from_file(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
    
    first_line = lines[0].strip().split()
    num_vertices = int(first_line[0])
    num_edges = int(first_line[1])
    
    graph = defaultdict(list)
    
    for line in lines[1:]:
        vertex1, vertex2, weight = line.strip().split()
        weight = int(weight)
        graph[vertex1].append((vertex2, weight))
        graph[vertex2].append((vertex1, weight))  # If the graph is undirected
    
    return graph

# BFS to find path from start to goal
def bfs_path(graph, start, goal):
    queue = deque([(start, [start])])
    visited = set()
    
    while queue:
        vertex, path = queue.popleft()
        if vertex in visited:
            continue
        visited.add(vertex)
        
        for neighbor, _ in graph[vertex]:
            if neighbor == goal:
                return path + [neighbor]
            queue.append((neighbor, path + [neighbor]))
    
    return None

# DFS to find path from start to goal
def dfs_path(graph, start, goal):
    stack = [(start, [start])]
    visited = set()
    
    while stack:
        vertex, path = stack.pop()
        if vertex in visited:
            continue
        visited.add(vertex)
        
        for neighbor, _ in graph[vertex]:
            if neighbor == goal:
                return path + [neighbor]
            stack.append((neighbor, path + [neighbor]))
    
    return None

# Best-First Search starting from 'start'
def best_first_search(graph, start):
    priority_queue = [(0, start, [start])]
    visited = set()
    result_order = []

    while priority_queue:
        priority_queue.sort()  # Sort the list to simulate a priority queue
        weight, vertex, path = priority_queue.pop(0)  # Pop the smallest element
        if vertex in visited:
            continue
        visited.add(vertex)
        result_order.append(vertex)

        for neighbor, weight in graph[vertex]:
            if neighbor not in visited:
                priority_queue.append((weight, neighbor, path + [neighbor]))

    return result_order

# Read graph from file
file_path = 'graph.txt'
graph = read_graph_from_file(file_path)

# Perform BFS from A to F
bfs_result = bfs_path(graph, 'A', 'F')
print("\nBFS từ đỉnh A đến F:")
print(bfs_result)

# Perform DFS from S to G
dfs_result = dfs_path(graph, 'S', 'G')
print("\nDFS từ đỉnh S đến G:")
print(dfs_result)

# Perform Best-First Search from S
best_first_search_result = best_first_search(graph, 'S')
print("\nBest-First Search từ đỉnh S:")
print(best_first_search_result)
```
