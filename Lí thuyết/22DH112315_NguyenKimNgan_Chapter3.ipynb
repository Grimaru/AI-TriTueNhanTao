```python
def unifcost(start, goal):
    OPEN = [(start, 0)]  # Tuple of (node, distance from start)
    CLOSED = []
    PREDECESSOR = {}  # Initialize PREDECESSOR and FVALUE dictionaries
    FVALUE = {}
    PREDECESSOR[start] = None
    FVALUE[start] = 0
    COUNT = 0  # Initialize COUNT
    DISTANCE = {
        'A': {'B': 1, 'C': 4},
        'B': {'C': 2, 'D': 3},
        'C': {'E': 5},
        'D': {'F': 2, 'G': 4},
        'E': {'G': 3},
        'F': {'G': 1},
        'G': {}
    }
    FN = {
        'A': 5, 'B': 4, 'C': 4, 'D': 3,
        'E': 7, 'F': 4, 'G': 2,
    }
    steps = []  # Store steps
    while OPEN != []:
        step_info = {'OPEN': [(node, dist) for (node, dist) in OPEN], 'CLOSED': CLOSED.copy()}
        steps.append(step_info)
        n, n_dist = deleteMin(OPEN, FVALUE, goal)  # Passing goal to stop if goal found
        CLOSED.append(n)
        COUNT += 1
        if n == goal:
            return extractPath(n, PREDECESSOR), FVALUE, steps  # Return path, FVALUE, and steps if goal found
        lst = successors(n, DISTANCE)  # Pass DISTANCE to get successors
        lst = listDifference(lst, CLOSED)
        for elt in lst:
            temp = FVALUE[n] + DISTANCE[n][elt] + FN[elt]  # Combine distance and heuristic
            if elt in [node for (node, dist) in OPEN]:
                existing_dist = [dist for (node, dist) in OPEN if node == elt][0]
                if temp < existing_dist:
                    FVALUE[elt] = temp - FN[elt]  # Update FVALUE without heuristic
                    PREDECESSOR[elt] = n
                    OPEN.remove((elt, existing_dist))
                    OPEN = insert((elt, temp), OPEN)  # Pass (node, distance) to insert
            else:
                FVALUE[elt] = temp - FN[elt]  # Update FVALUE without heuristic
                PREDECESSOR[elt] = n
                OPEN = insert((elt, temp), OPEN)  # Pass (node, distance) to insert
    return None, FVALUE, steps  # If no path found, return None, FVALUE, and steps

def deleteMin(lst, fn, goal):
    minVal = float('inf')
    minElt = None
    minDist = None
    for elt, dist in lst:
        temp = fn[elt]
        if temp < minVal:
            minVal = temp
            minElt = elt
            minDist = dist
    lst.remove((minElt, minDist))
    if minElt == goal:  # Stop if goal found
        lst.clear()
    return minElt, minDist

def insert(elt, lst):
    if lst == []:
        return [elt]
    for i, (node, dist) in enumerate(lst):
        if elt[1] < dist:
            return lst[:i] + [elt] + lst[i:]
    return lst + [elt]

def successors(n, DISTANCE):
    if n in DISTANCE:
        return list(DISTANCE[n].keys())
    else:
        return []

def listDifference(lst1, lst2):
    return [x for x in lst1 if x not in lst2]

def extractPath(goal, PREDECESSOR):
    path = [goal]
    while PREDECESSOR[goal] is not None:
        goal = PREDECESSOR[goal]
        path.append(goal)
    return path[::-1]

def test():
    start = 'A'
    goal = 'G'
    #print("Starting a uniform-cost graph search from " + start)
    path, FVALUE, steps = unifcost(start, goal)
    if path:
        print("Path found:", path)
        print(str(len(path) - 1) + " nodes expanded during the search.")
        print('Total distance is', FVALUE[goal])
    else:
        print("Goal not reachable.")

    print("Steps:")
    for i, step in enumerate(steps):
        print("Step", i + 1)
        print("OPEN:", step['OPEN'])
        print("CLOSED:", step['CLOSED'])
        print()

test()
```
