```python
#BÃ i 2
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

class DataClustering:
    def __init__(self, file_path):
        self.df = pd.read_csv(file_path)
    
    def clustering_centers(self):
        kmeans = KMeans(n_clusters=3)
        self.df['Cluster'] = kmeans.fit_predict(self.df[['ApplicantIncome', 'LoanAmount']])
        return kmeans.cluster_centers_
    
    def plot_clusters(self):
        kmeans = KMeans(n_clusters=3)
        self.df['Cluster'] = kmeans.fit_predict(self.df[['ApplicantIncome', 'LoanAmount']])
        
        plt.figure(figsize=(15, 5))
        for cluster_id in range(3):
            cluster = self.df[self.df['Cluster'] == cluster_id]
            plt.scatter(cluster['ApplicantIncome'], cluster['LoanAmount'], label=f'Cluster {cluster_id + 1}')
        
        plt.xlabel('ApplicantIncome')
        plt.ylabel('LoanAmount')
        plt.title('Clusters')
        plt.legend()
        plt.show()

# Usage Example
data_clustering = DataClustering('data1.csv')
print("Cluster Centers:", data_clustering.clustering_centers())
data_clustering.plot_clusters()
```
