```python
#Bài 1
import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu từ file CSV
df = pd.read_csv('data_information.csv')

# Bước 2
# Hiển thị 5 dòng đầu
print("5 dòng đầu:")
print(df.head())

# Hiển thị 5 dòng cuối
print("\n5 dòng cuối:")
print(df.tail())

# Bước 3
print("\nTên các cột:")
print(df.columns)

# Bước 4
print("\nSố dòng, số cột:")
print(df.shape)

# Bước 5
# Thông tin mô tả
print("\nThông tin mô tả:")
print(df.describe())

# Kiểu dữ liệu từng cột
print("\nKiểu dữ liệu từng cột:")
print(df.dtypes)

# Bước 6
duplicate_age = df[df.duplicated(['age'])]
print("\nDữ liệu trùng lặp trong cột 'age':")
print(duplicate_age)

# Bước 7
unique_age_df = df.drop_duplicates(subset=['age'])
print("\nData sau khi bỏ trùng lặp trong cột 'age':")
print(unique_age_df)

# Số dòng còn lại
print("\nSố dòng còn lại:")
print(unique_age_df.shape[0])
```

```python
# Bước 8
df.plot()
plt.show()

# Bước 9
df.plot(x='name', y='age', kind='scatter')
plt.show()

df.plot(x='point', y='age', kind='line')
plt.show()
```

```python
# Bước 10 + Bước 11
# Tạo DataFrame chỉ gồm cột name và age
df_name_age = df[['name', 'age']]
y1 = df_name_age['age'].values
plt.subplot(1, 2, 1)
plt.title('Chart 1')
plt.plot(y1, linestyle='-', color='blue')

# Tạo DataFrame chỉ gồm cột age và point
df_age_point = df[['age', 'point']]
x2 = df_age_point['point'].values
y2 = df_age_point['age'].values
plt.subplot(1, 2, 2)
plt.title('Chart 2')
plt.plot(x2, linestyle='-', color='green')
plt.plot(y2, linestyle='-', color='red')
```

```python
# Bài 2
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Bước 1
who_covid_data=pd.read_csv('WHO-COVID-19-global.csv')
headers=list(who_covid_data)
print(headers)
```

```python
# Bước 2
every_nth_days=1
country_df= who_covid_data[who_covid_data[' Country']=='Brazil']
info=['cases']
for info_type in info:
    print(info_type)
    brazil_df=pd.DataFrame(country_df)
    print(brazil_df.head(10))
    brazil_df.plot()
    plt.show()
```

```python
# Bước 3
every_nth_days=1
country_df= who_covid_data[who_covid_data[' Country']=='Brazil']
info=['cases']
for info_type in info:
    fig, ax1=plt.subplots(figsize=(20,10), dpi=50)
    color='tab:blue'
    ax1.set_xlabel('days', fontsize='large',fontweight='bold')
    ax1.bar(country_df['Date_reported'][::every_nth_days],country_df[' New_{}'.format(info_type)][::every_nth_days],color=color,linewidth=20,label='New {}'.format(info_type))
    ax1.tick_params(axis='y', labelcolor=color, labelsize='large')
    #bo sung
    ax2=ax1.twinx()
    
    color='tab:red'
    ax2.plot(country_df['Date_reported'][::every_nth_days],country_df[' Cumulative_{}'.format(info_type)][::every_nth_days], color=color, label='Cumulative {}'.format(info_type), linewidth=3)    
    ax2.tick_params(axis='y', labelcolor=color, labelsize=20)    
    #end bo sung
    for idx, t in enumerate (ax1.get_xticklabels()):
        t.set_fontsize(11)
        t.set_fontweight('bold')
        if (idx % 5)!=0:
            t.set_visible(False)
    plt.ticklabel_format(style='plain',axis='y')
    plt.xticks(rotation=90)
    dates=country_df['Date_reported'].unique()
    start_date=dates[0]
    end_date=dates[len(dates)-1]
    
    plt.title('{} (from {} to {})'.format('Brazil',start_date, end_date),fontsize=30,fontweight='bold')
    fig.tight_layout()
    ax1.legend(loc='upper left',fontsize=20)
    ax2.legend(loc='upper right',fontsize=20)
    plt.setp(ax1.xaxis.get_majorticklabels(),rotation=45)
    
    plt.show()
```
