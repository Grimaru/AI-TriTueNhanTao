```python
#Bài 1
class Graph:
    def __init__(self, adjac_lis):
        self.adjac_lis = adjac_lis

    def get_neighbors(self, v):
        return self.adjac_lis[v]
    
    def h (self, n):
        H = {
            'A': 1,
            'B': 1,
            'C': 1,
            'D': 1
        }

        return H[n]
    
    def a_star_algorithm(self, start, stop):
        open_lst = set([start])
        close_lst = set([])

        p = {}
        p[start] = 0

        par = {}
        par[start] = start

        while len(open_lst) > 0:
            n = None

            for v in open_lst:
                if n == None or p[v] + self.h(v) < p[n] + self.h(n):
                    n = v

            if n == None:
                print('Path does not exist!')
                return None
            
            elif n == stop:
                reconst_path = []
                total_weight = p[n]

                while par[n] != n:
                    reconst_path.append(n)
                    n = par[n]

                reconst_path.append(start)
                reconst_path.reverse()

                print('Path found: {}'.format(reconst_path))
                print('Total weight: {}'.format(total_weight))
                return reconst_path, total_weight
            
            for (m, weight) in self.get_neighbors(n):
                if m not in open_lst and m not in close_lst:
                    open_lst.add(m)
                    par[m] = n
                    p[m] = p[n] + weight

                elif p[m] > p[n] + weight:
                    p[m] = p[n] + weight
                    par[m] = n

                    if m in close_lst:
                        close_lst.remove(m)
                        open_lst.add(m)

            open_lst.remove(n)
            close_lst.add(n)

        print('Path does not exist!')
        return None
```

```python
adjac_lis = {
    'A': [('B', 5), ('C', 1), ('D', 10)],
    'B': [('D', 5)],
    'C': [('D', 5)]
}
graph1 = Graph(adjac_lis)
graph1.a_star_algorithm('A', 'D')
```

```python
#Bài 3
class Water:
    def __init__(self, adjac_lis):
        self.adjac_lis = adjac_lis

    def get_neighbors(self, v):
        return self.adjac_lis[v]
    
    def h(self, n):
        H = {
            'A': 1,
            'B': 1,
            'C': 1,
            'D': 1
        }
        return H[n]
    
    def a_star_algorithm(self, start, stop):
        open_lst = set([start])
        close_lst = set([])

        p = {}
        p[start] = 0

        par = {}
        par[start] = start

        while len(open_lst) > 0:
            n = None

            for v in open_lst:
                if n is None or p[v] + self.h(v) < p[n] + self.h(n):
                    n = v

            if n is None:
                print('Path does not exist!')
                return None
            
            elif n == stop:
                reconst_path = []
                total_weight = p[n]

                while par[n] != n:
                    reconst_path.append(n)
                    n = par[n]

                reconst_path.append(start)
                reconst_path.reverse()

                print(f'Path found: {reconst_path} with total weight: {total_weight}')
                return reconst_path, total_weight
            
            for (m, weight) in self.get_neighbors(n):
                if m not in open_lst and m not in close_lst:
                    open_lst.add(m)
                    par[m] = n
                    p[m] = p[n] + weight

                elif p[m] > p[n] + weight:
                    p[m] = p[n] + weight
                    par[m] = n

                    if m in close_lst:
                        close_lst.remove(m)
                        open_lst.add(m)

            open_lst.remove(n)
            close_lst.add(n)

        print('Path does not exist!')
        return None

def read_graph_from_file(file_path):
    adjac_lis = {}
    with open(file_path, 'r') as file:
        for line in file:
            parts = line.strip().split()
            if len(parts) == 3:
                start, end, weight = parts[0], parts[1], int(parts[2])
                if start not in adjac_lis:
                    adjac_lis[start] = []
                if end not in adjac_lis:
                    adjac_lis[end] = []
                adjac_lis[start].append((end, weight))
                adjac_lis[end].append((start, weight))
    return adjac_lis
```
